# Figure 5: Sign-conditioned direct finite-N dynamical test

This notebook performs direct finite-\(N\) dynamical tests on
hyperspheres, product manifolds, and non-spherical representatives
from Table VI. For zero-Euler-characteristic controls we use
ordinary drift ensembles. For nonzero-Euler-characteristic cases
we use structurally prescribed sign-conditioned drift ensembles.
All generated samples are retained. The dynamical sweeps are
performed with logarithmic \(K/K_{\rm ref}\) grids.

```python
from pathlib import Path
import sys

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from matplotlib.lines import Line2D

# Set the experiment root for runs started from the notebook directory or the
# experiment directory.
HERE = Path.cwd().resolve()
EXPERIMENT_ROOT = HERE.parent if HERE.name == "notebooks" else HERE
SRC_DIR = EXPERIMENT_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

from paths import DATA_DIR, FIGURE_DIR, ensure_directories
from plotting import set_paper_style, save_figure
from manifold_sync import *

ensure_directories()
set_paper_style()


def manuscript_palette(labels):
    # Return the blue-green manuscript palette used for dimension-like groups.
    labels = list(labels)
    colors = sns.color_palette("viridis", n_colors=len(labels))
    return dict(zip(labels, colors))
```

```python
from finite_oscillator_dynamics import (
    DATA_DIR,
    FIGURE_DIR,
    load_or_run,
    make_figure,
)
import shutil

RECOMPUTE = False
(
    raw,
    summary,
    fit_summary,
    construction,
    merged,
    high,
    product_raw,
    product_summary,
    tablevi_raw,
    tablevi_summary,
) = load_or_run(recompute=RECOMPUTE)
fig = make_figure(summary, high, product_raw, product_summary, tablevi_raw, tablevi_summary)

source = FIGURE_DIR / "fig05_direct_dynamics_600dpi.png"
manuscript_root = EXPERIMENT_ROOT.parent
shutil.copyfile(source, manuscript_root / "Fig5.png")
shutil.copyfile(source, manuscript_root / "latex" / "Fig5.png")

print("Final samples by manifold dimension:")
display(
    summary.groupby(["manifold_dimension", "chi"]).agg(
        samples=("jump_size", "size"),
        jump_mean=("jump_size", "mean"),
        scan_difference_mean=("max_scan_difference", "mean"),
    )
)
print("Ensemble construction summary:")
display(
    construction.groupby(["manifold_dimension", "chi", "construction_status"]).agg(
        samples=("seed", "size"),
    )
)
print("Product manifold samples:")
display(
    product_summary.groupby(["case", "chi", "role"]).agg(
        samples=("jump_size", "size"),
        jump_mean=("jump_size", "mean"),
        scan_difference_mean=("max_scan_difference", "mean"),
    )
)
print("Table VI non-spherical samples:")
display(
    tablevi_summary.groupby(["case", "chi", "role"]).agg(
        samples=("jump_size", "size"),
        jump_mean=("jump_size", "mean"),
        scan_difference_mean=("max_scan_difference", "mean"),
    )
)
```
